A primeira parte consiste em adiquirir o data set.

```python
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
arq = "https://raw.githubusercontent.com/igordrodrigues/workshop01122022/main/breast-cancer.csv"
base = np.loadtxt(arq, delimiter=',', dtype=str)

print(base, len(base))
print(base[0])
```

Em seguida precisamos transformar o arquivo em um formato que possamos manipular. Vamos aproveitar e remover instancias com informação de alguma feature faltando.

```python
import numpy as np
z =[]
eliminador = 0
for line in range(0, len(base) - 1):
    try:
        aux = base[line]
        aux2 = []
        for i in range(0,len(base[line])):
          aux2.append(int(base[line][i]))
        z.append(aux2)
    except:
        eliminador =eliminador +1
print(eliminador)
```

Em seguida, vamos verificar quantas instancias foram eliminadas, e conferir como nosso dataset ficou.

```python
print('eliminados: ',eliminador)
base =z
print('tamanho do dataset: ',len(base))
print('primeira instancia do dataset: ', base[1])
print('numero de features por instancia: ',len(base[1]))
```

No nosso dataset a classificação esta no junto dos outros atributos, dessa forma iremos precisar separalos, para que o algoritmo não tenha acesso a essa informação quando estiver realizando a classificação de novas amostras. Nesse caso especifico, a classificação é o primeiro elemento de cada linha.

```python
X = []
Y = []
for linha in base:
    subject = []
    for i in range(1,len(linha)-1):
      subject.append(linha[i])
    X.append(subject)
    Y.append(linha[0])
print(X)
print(Y)
```

Muito bem, temos os atributos no vetor X e as suas respectivas classificações no vetor Y, onde X[i] esta classificado em Y[i].
Agpra precisamos dividir nossa amostra em teste e treino, vamos lá. Usaremos uma função da biblioteca sickit learn para fazer essa divisão.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.35,random_state=120)
```

Agora temos duas amostras, uma que sera usada para treinar nosso modelo e outra para testa-lo.
É hora da diversão, podemos começar a treinar e classificar.

```python
from sklearn import svm

model = svm.SVC(C=1,gamma='auto', kernel='linear', class_weight='balanced', degree=3, max_iter=2000000)
model.fit(X_train, y_train)
svm_pred = model.predict(X_test)
acc = model.score(X_test, y_test)
tp = 0
tn = 0
fp = 0
fn = 0
for j in range(len(svm_pred)):
  if svm_pred[j] == 0:
    if y_test[j] == 0:
       tn = tn + 1
    else:
      fn = fn + 1
  else:
    if y_test[j] == 1:
       tp = tp + 1
    else:
       fp = fp + 1
sns = tp / (tp + fn)
spc = tn / (tn + fp)
print(acc,sns,spc)
```

```python
y_pred = model.predict(X_test)
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred))
print("ROC-AUC: ",metrics.roc_auc_score(y_test, y_pred))
print("f1 score: ", metrics.f1_score(y_test, y_pred, average='weighted'))
```

Podemos tentar melhorar ainda mais nosso modelo. Uma alternativa é utilizar conhecimentos sobre a amostra e tentar pre-selecionar features que possam possuir maior impacto na classificação. A seguir iremos selecionar algumas features e testar o resultado.
